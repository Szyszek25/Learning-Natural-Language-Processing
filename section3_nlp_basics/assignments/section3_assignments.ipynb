## 1. Text Preprocessing with Pandas

1. Read the _childrens_books.csv_ file into a Jupyter Notebook
2. Within the Description column:
* Make all the text lowercase
* Remove all \xa0 characters - spacja nieprzełamująca
* Remove all punctuation

```python
import pandas as pd

pd.set_option('display.max_colwidth', None)
```

```python
df = pd.read_csv('../data/childrens_books.csv')
df.head()
```

```python
df['Description_Clean']=df.Description.str.lower() #można tak i tak
df.head()
```

```python
df.Description.iloc[0] #wiersz
```

```python
df['Description_Clean']=df['Description_Clean'].str.replace('\xa0',' ') #jeśli usunę znak i jeszcze raz uruchomię, to nie wstawię drugi raz znaku
df.head()
```

```python
df.Description_Clean.iloc[0] #wiersz
```

```python
df['Description_Clean']=df['Description_Clean'].str.replace(r'[^\w\s]','', regex=True) #usuniecie interpunkcji
df.head()
```

## 2. Text Preprocessing with spaCy

In addition to the lowercasing and special character removal from the previous assignment, within the cleaned Description column:
* Tokenize the text
* Lemmatize the text
* Remove stop words

```python
import spacy
nlp = spacy.load('en_core_web_sm')
import en_core_web_sm
nlp = en_core_web_sm.load()
```

```python
def token_lemma_nonstop(text):
    doc=nlp(text)
    output=[token.lemma_ for token in doc if not token.is_stop] #lista lematyzowanych tokenów
    output=' '.join(output) 
    return output
```

```python
df['Description_Clean']=df.Description_Clean.apply(token_lemma_nonstop)
df.head()
```

## 3. Count Vectorizer

1. Vectorize the cleaned and normalized text using Count Vectorizer with the default parameters
2. Modify the Count Vectorizer parameters to reduce the number of columns:
* Remove stop words
* Set a minimum document frequency of 10%
3. Use the updated Count Vectorizer to identify the:
* Top 10 most common terms
* Top 10 least common terms that appear in at least 10% of the documents
4. Create a horizontal bar chart of the top 10 most common terms

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(stop_words='english',min_df=0.1)
```

```python
dtm=cv.fit_transform(df['Description_Clean'])
dtm_df=pd.DataFrame(dtm.toarray(), columns=cv.get_feature_names_out())
dtm_df
```

```python
most_common_terms=dtm_df.sum()
most_common_terms
```

```python
most_common_terms.sort_values().tail(10).plot(kind='barh')
```

```python
most_common_terms.nlargest(10).sort_values().plot(kind='barh');
```

```python
#bottom
bottom_terms=dtm_df.sum()
```

```python
bottom_terms=bottom_terms.sort_values().head(10).plot(kind='barh') #sortowanie rosnące, czyli najmniejsze wartosci to głowa
```

## 4. TF-IDF Vectorizer

1. Vectorize the cleaned and normalized text using TF-IDF Vectorizer with the default parameters
2. Modify the TF-IDF Vectorizer parameters to reduce the number of columns:
* Remove stop words
* Set a minimum document frequency of 10%
* Set a maximum document frequency of 50%
3. Using the updated TF-IDF Vectorizer, create a  horizontal bar chart of the top 10 most highly weighted terms
4. Compare the Count Vectorizer bar chart from the previous assignment with the TF-IDF Vectorizer bar chart and note the differences in the top term lists

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tv=TfidfVectorizer(stop_words='english',min_df=0.1,max_df=0.5)
tfidf=tv.fit_transform(df['Description_Clean'])
tfidf_df=pd.DataFrame(tfidf.toarray(),columns=tv.get_feature_names_out())
tfidf_df
```

```python
top10_weights=tfidf_df.sum()
top10_weights=top10_weights.sort_values().tail(10).plot(kind='barh')
```

```python
last10_weights=tfidf_df.sum()
last10_weights.sort_values().head(10).plot(kind='barh')
```

```python
import matplotlib.pyplot as plt


fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))


most_common_terms.nlargest(10).sort_values().plot(kind='barh', ax=axes[0], color='skyblue')
axes[0].set_title('Count Vectorizer')
axes[0].invert_yaxis()


top10_weights.plot(kind='barh', ax=axes[1], color='salmon')
axes[1].set_title('TF-IDF Vectorizer')
axes[1].invert_yaxis()

plt.tight_layout()
plt.show()
```
